#suppose a dataset has a Gender column with categorical elements like Male and  Female.  These labels have no specific order of preference and also since the data is string labels, machine learning models misinterpreted that there is some sort of hierarchy in them

#One way to solve this is by encoding where we will assign a numerical value to these labels for example Male and Female mapped to 0 and 1.
This methos is called One Hot Encoding technique

```python
#Let us see this by example
import pandas as pd
df = pd.read_csv("/content/sample_data/employee_data All in one.csv")
df.head()
```

```python
df.nunique()
```

```python
df["GenderCode"].value_counts()
df["EmployeeStatus"].value_counts()
```

```python
df.nunique()
```

```python
# one_hot_encoded_data = pd.get_dummies(data, columns = ['Remarks', 'Gender'])
#print(one_hot_encoded_data)
df1 = pd.get_dummies(df, columns = ["EmployeeStatus", "EmployeeType", "PayZone", "EmployeeClassificationType", "TerminationType", "DepartmentType", "GenderCode", "RaceDesc", "MaritalDesc", "Performance Score", "Current Employee Rating"])
df1
```

#Another way of one hot encoding by using Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
df4 = pd.read_csv("/content/sample_data/employee_data All in one.csv")
df4
```

```python
df4.nunique()
```

```python
df4["GenderCode"] = df4["GenderCode"].astype("category")
df4["RaceDesc"] = df4["RaceDesc"].astype("category")
df4["G_Code"] = df4["GenderCode"].cat.codes
df4["R_Desc"] = df4["RaceDesc"].cat.codes
encode = OneHotEncoder()
df40 = pd.DataFrame(encode.fit_transform(df4[["G_Code", "R_Desc"]]).toarray())
df5 = df4.join(df40)
df5
```
