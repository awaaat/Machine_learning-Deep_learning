```python
import os
import numpy as np
import tensorflow as tf
import tensorflow_hub as hub
import shutil
import re
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
```

```python
#Let us check the versions

print(f"version: {tf.__version__}")
print(f"tf hub version: {hub.__version__}")
print(f"eager mode: {tf.executing_eagerly()}")
print('GPU is', 'available' if tf.config.list_logical_devices("GPU") else 'not available')
```

#Let us Get the Dataset

```python
train_data, validation_data, test_data = tfds.load(
    name = "imdb_reviews",
    split = ("train[:60%]", "train[60%:]", "test"),
    as_supervised = True
)
```

```python
train_examples_batch, train_labels_batch = next(iter(train_data.batch(10)))
train_examples_batch
```

*italicized text*# We now buils our model

```python
embeding = "https://tfhub.dev/google/nnlm-en-dim50/2"
tf_hub_layer = hub.KerasLayer(embeding, input_shape  = [], dtype = tf.string, trainable = True)
tf_hub_layer(train_examples_batch[:3])
```

```python
#Model Development
model = tf.keras.Sequential()
model.add(tf_hub_layer)
model.add(tf.keras.layers.Dense(16, activation = "relu"))
model.add(tf.keras.layers.Dense(1))


model.summary()
```

#Picking a loss function and an optimizer

```python
model.compile(optimizer = "adam", loss = tf.keras.losses.BinaryCrossentropy(from_logits = True), metrics = ["accuracy"])
```

```python
#Let us now train the model
history  = model.fit(train_data.shuffle(10000).batch(512), validation_data = validation_data.batch(512), verbose = 1)
```

```python
#Model Evaluation
results = model.evaluate(test_data.batch(512), verbose = 1)
for name, value in zip(model.metrics_names, results):
  print("%s: %.3f" % (name, value))
```
