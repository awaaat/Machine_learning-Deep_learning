```python
from sklearn.datasets import make_blobs
from matplotlib import style
import matplotlib.pyplot as plt
```

```python
style.use("fivethirtyeight")
X, y = make_blobs(n_samples = 1000, n_features = 2, shuffle=True, random_state = 40, cluster_std=5, centers=5)
plt.scatter(X[:, 0], X[:, 1], s = 40, color = "g")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
X, y = make_moons(n_samples = 1000, shuffle = True, noise = 0.003, random_state = 42)
plt.scatter(X[:, 0], X[:, 1], s = 40, c  = "b")

plt.ylabel("y")
plt.xlabel("X")

plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
X, y = make_circles(n_samples = 1000, shuffle = True, noise = 0.000005, random_state = 40)
plt.scatter(X[:, 0], X[:, 1], s = 40, c = "g" )
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
plt.clf()
```

```python
from sklearn.datasets import make_multilabel_classification
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = make_multilabel_classification(n_samples = 1000,
                                      n_features = 4,
                                      n_classes = 2, n_labels = 1,
                                      allow_unlabeled = True,
                                      length = 50,
                                      random_state = 40)
df = pd.concat([pd.DataFrame(X,columns =["X1", "X2", "X3", "X4"]), pd.DataFrame(y, columns =["Label1", "Label2"])], axis = 1)
display(df.head(4))

plt.scatter(df["X1"], df["X2"], c = df["Label1"])
plt.show()
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples = 100, n_features = 1, n_informative = 2, n_targets = 1, bias = 0.02, noise = 0.03, random_state = 40, shuffle = True)
plt.scatter(X, y)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
plt.legend
```

```python
from sklearn.datasets import make_sparse_uncorrelated
import matplotlib.pyplot as plt
X, y = make_sparse_uncorrelated(n_samples  = 100, n_features = 4, random_state  = 40)
for i in range(4):
  plt.subplot(2,2, i+1)
  plt.scatter(X[:,i], y)
  plt.xlabel("X" + str(i+1))
  plt.ylabel("y")
plt.show()
```

```python
from sklearn.datasets import make_friedman2
X, y = make_friedman2(n_samples = 500, noise = 0.03, random_state = 42)
import matplotlib.pyplot as plt
for i in range(4):
  plt.subplot(2,2, i+1)
  plt.scatter(X[:, i], y)
  plt.xlabel("X" + str(i+1))
  plt.ylabel("y")
plt.show()
```
