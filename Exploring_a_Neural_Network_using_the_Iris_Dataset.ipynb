```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier

# Load the iris dataset
iris = load_iris()

# look at the four (4) features
pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)
```

```python
# Create a neural network
#  hidden_layer_sizes=(10,5,15) -- this would mean there are 3 layers, with the number of neurons in layer 1 at 10, layer 2 at 5, and layer 3 at 15
clf = MLPClassifier(hidden_layer_sizes=(10), activation='relu', solver='adam', max_iter=1000)

# Fit the neural network to the iris dataset
clf.fit(X_train, y_train)
```

```python
# output all of the weights for every node connection, with the first index as layer, second index as neuron within layer
for i in range(len(clf.coefs_)):
    number_neurons_in_layer = clf.coefs_[i].shape[1]
    for j in range(number_neurons_in_layer):
        weights = clf.coefs_[i][:,j]
        print(str(i)+", "+str(j)+": ", weights, end=", ")
        print()
    print()
```

```python
# calculate the TEST accuracy
accuracy_score(y_test, clf.predict(X_test))
```
