```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions
```

```python
# Load dataset
iris = datasets.load_iris(as_frame=True)
X = iris.data[["sepal length (cm)", "sepal width (cm)"]] # we only take two features for easy visualization
y = iris.target # 0,1,2 --> species of iris flower

X.join(y)
```

```python
def knn_comparison(data, k):
  x = data[["sepal length (cm)", "sepal width (cm)"]].values # we only take these two features for easy visualization
  y = data['target'].astype(int).values
  clf = neighbors.KNeighborsClassifier(n_neighbors=k)
  clf.fit(x, y) # Plotting decision region

  # Adding axes annotations
  plot_decision_regions(x, y, clf=clf, legend=2)
  plt.xlabel("sepal length (cm)")
  plt.ylabel("sepal width (cm)")
  plt.title('KNN with K=' + str(k))
  plt.show()
```

```python
k = 1  # set the hyperparameter k

knn_comparison(X.join(y), k)
```
